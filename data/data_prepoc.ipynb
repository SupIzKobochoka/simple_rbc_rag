```python
import pandas as pd

df = pd.read_excel('rbc_2000_2020.xlsx')
df['publish_date'] = pd.to_datetime(df['publish_date'], utc=True).dt.strftime('%Y-%m-%d')
df = df[['publish_date', 'text', 'fronturl']].dropna()
df.drop(df[df['text'].str.split().str.len() == 0].index, axis=0, inplace=True)
display(df.head(2))
df.to_csv('data.csv', index=False)
```
